# 오존농도시각화

```python
import pandas as pd
df = pd.read_csv('./data/년도별오존지수_경기도.csv', engine='python', encoding='CP949')
df
```

```python
df.columns=['연도','오존']
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

##########한글폰트 적용
path = 'C:/Windows/Fonts/gulim.ttc'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)
```

```python
# plt.xticks(size=10, rotation=45)
# plt.plot(df['연도'], df['오존'], marker='o', markersize=5)
```

```python
from matplotlib.ticker import MaxNLocator

ax = plt.figure().gca()

df.groupby(["연도"])["오존"].mean().plot()

ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.xticks(size=10, rotation=45)

plt.title('경기 오존농도', size=20) 
plt.plot(df['연도'], df['오존'], marker='o', markersize=5)
```

# 운영형태별 상품군별 거래액 시각화

```python
import pandas as pd
df1 = pd.read_csv('./data/온라인쇼핑몰_운영형태별_상품군별거래액_20210318133241.csv', engine='python', encoding='CP949')
df1.head()
```

```python
df1["유형"] = df1["상품군별"] + " " + df1["운영형태별"]
df1
```

```python
df1.drop(['상품군별','운영형태별'],axis=1,inplace=True)
```

```python
df1.set_index('유형', inplace=True)
```

```python
df1
```

```python
df1_pp = df1.T
df1_pp
```

```python
# 데이터프레임의 인덱스를 정수형으로 변경 (x축 눈금 라벨 표시)
df1_pp.index = df1_pp.index.astype(int)
df1_pp.index
```

```python
##########한글폰트 적용
path = 'C:/Windows/Fonts/gulim.ttc'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)
```

```python

# 막대 그래프 그리기
df1_pp.plot(kind='bar',figsize=(10, 5), width=0.7, color=['#ddb6f2', 'green', 'skyblue', '#abf28a'])


plt.title('상품군/운영형태별 거래액', size=20)
plt.xticks(size=10, rotation=0)
plt.yticks(size=10, rotation=0)

plt.ylabel('거래액', size=10)
plt.xlabel('기간', size=10)
plt.ylim(5000, 9000000)
plt.legend(loc='best', fontsize=10)

plt.show()
```
